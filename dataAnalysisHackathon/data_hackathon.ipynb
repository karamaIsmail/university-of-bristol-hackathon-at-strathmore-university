```python
%matplotlib inline 
import pandas as pd
#import mpld3
#mpld3.enable_notebook()
import matplotlib
import matplotlib.pyplot as plt 
import squarify
```

```python
data = pd.read_excel("hypertensionSmoking.xlsx",sheet_name=1) 
data 
```

```python
data.rename(columns={"CCG Name":"CCG_Name"})
```

```python
df = data.iloc[:,0:16].drop(columns=["Hypertension: QOF prevalence (all ages).1",
                                "Hypertension: QOF prevalence (all ages).2",
                                "Hypertension: QOF prevalence (all ages).4",
                                "Hypertension: QOF prevalence (all ages).5",
                                "Hypertension: QOF prevalence (all ages).7",
                                "Hypertension: QOF prevalence (all ages).8",
                                "Hypertension: QOF prevalence (all ages).10",
                                "Hypertension: QOF prevalence (all ages).11",
                                "Hypertension: QOF prevalence (all ages).13",
                                "Hypertension: QOF prevalence (all ages).14"], axis=1)\
                      .rename(columns={"CCG Name":"CCG",
                                     "Hypertension: QOF prevalence (all ages)":"2009/10",
                                     "Hypertension: QOF prevalence (all ages).3":"2010/11",
                                     "Hypertension: QOF prevalence (all ages).6":"2011/12",
                                     "Hypertension: QOF prevalence (all ages).9":"2012/13",
                                     "Hypertension: QOF prevalence (all ages).12":"2013/14"})\
                      .drop(index=[0,1], axis=0)\

df
```

```python
color = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')
df.plot.box(vert=False, color=color, sym="r+", figsize=[14,6])
plt.title("Hypertension prevalence for all ages")


print ("The mean for 2009/10 is %.3f" %df["2009/10"].mean())
print ("The mean for 2010/11 is %.3f" %df["2010/11"].mean())
print ("The mean for 2011/12 is %.3f" %df["2011/12"].mean())
print ("The mean for 2012/13 is %.3f" %df["2012/13"].mean())
print ("The mean for 2013/14 is %.3f" %df["2013/14"].mean()) 

print ("=================================")
print ("The median for 2009/10 is %.3f" %df["2009/10"].median())
print ("The median for 2010/11 is %.3f" %df["2010/11"].median())
print ("The median for 2011/12 is %.3f" %df["2011/12"].median())
print ("The median for 2012/13 is %.3f" %df["2012/13"].median())
print ("The median for 2013/14 is %.3f" %df["2013/14"].median()) 
```
